```python
''' 2. An electrical manufacturing company conducts quality checks at specified periods on the 
products it manufactures. Historically, the failure rate for the manufactured item is 5%. 
Suppose a random sample of 10 manufactured items is selected. Answer the following questions. [4 Marks]'''

# Since there are only two possible outcomes with defined success and failure prbabilities and the 
# trails are independent the random variable follows a binomial distribution.

# Let p be the  rate of success
p= 0.95
# Let n be the sample size
n= 10
k= np.arange(0,11)
binomial1= stats.binom.pmf(k,n,(1-p))
plt.plot(k, binomial1,'o-');
plt.title('Binomial plot');
plt.xlabel('Number of failures');
plt.ylabel('Probability of success');
```

```python
# 2.A. Probability that none of the items are defective? [1 Mark]
p_no_defective =  binomial1[0]
p_no_defective
```

```python
# 2.B. Probability that exactly one of the items is defective? [1 Mark]
p_one_defective = binomial1[1]
p_one_defective
```

```python
# 2.C. Probability that two or fewer of the items are defective? [1 Mark]
p_two_or_less= binomial1[2] + binomial1[1] + binomial1[0]
p_two_or_less
```

```python
# 2.D. Probability that three or more of the items are defective ? [1 Mark]
p_three_or_more = 1- (p_two_or_less)
p_three_or_more
```

```python
# 3. A car salesman sells on an average 3 cars per week. [3 Marks]

# Since only the average  value is known, the random variable follows a poisson distribution.

rate = 3
n = np.arange(0,10)
poisson= stats.poisson.pmf(n,rate)
plt.plot(n, poisson,'o-');
plt.title('Poisson distribution mass function');
plt.xlabel('Number of cars sold per week');
plt.ylabel('Probability of number of cars sold per week');
```

```python
# 3.A. What is Probability that in a given week he will sell some cars? [1 Mark]
p_some_cars = 1 - poisson[0]
p_some_cars
```

```python
# 3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]
p_two_or_more_less_than_five = poisson[2] + poisson[3] + poisson[4]
p_two_or_more_less_than_five
```

```python
# 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]
poisson2= stats.poisson.cdf(n,rate)
plt.plot(n,poisson2, 'o-')
plt.title('Cumulative probability distribution Poisson: $\lambda$ = %i ' % rate);
plt.xlabel('Number of cars sold per week');
plt.ylabel('cummulative probability of cars sold per week');
```

```python
''' 4. Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed,
marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures the
percentage of orders that are taken correctly is 86.8%. Suppose that you place an order with the bot and two friends of yours
independently place orders with the same bot. Answer the following questions. [3 Marks]'''

# The random variable follows a binomial distribution as the trials are independent 
# and have fixed probability of success and failure.

# Let p1 be the probability of success
p1=0.868

#Let n1 be the number of trials
n1 = 3

k1= np.arange(0,4,1)
binomial2= stats.binom.pmf(k1,n1,p1)
plt.plot(k1,binomial2,'--');
plt.title('Binomial distribution of speech based bot')
plt.xlabel('Number of successes')
plt.ylabel('Probability of success')
plt.show()
```

```python
# 4.B. What is the probability that none of the three orders will be recognised correctly? [1 Mark]
p_none_recognized = binomial2[0]
p_none_recognized
```

```python
# 4.C. What is the probability that at least two of the three orders will be recognised correctly? [1 Mark]
p_atleat_two = binomial2[2] + binomial2[3]
p_atleat_two
```

```python
from    scipy.stats             import  ttest_1samp,ttest_ind, wilcoxon, ttest_ind_from_stats
import  scipy.stats             as      stats  
from    statsmodels.stats.power import  ttest_power
from    statsmodels.stats.weightstats import ztest
from    statsmodels.stats.proportion import proportions_ztest
```

```python
# A. Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?
plt.figure(figsize=(5,5));
sns.barplot(data=startup_df, x='OperatingState', y='Funds_in_million');
plt.title('Operating State Vs Funds in Million bar plot');
```

```python
# B. Write the null hypothesis and alternative hypothesis.
```

$H_0$: The means of funds of companies still operating and closed down are the same.

$H_A$: The means of funds of companies still operating and closed down are not the same. 

```python
# C. test for significance and conclusion.
alpha= 0.05
new_cols=startup_df[['OperatingState','Funds_in_million']]
new_df=new_cols.copy()
new_df.head()
```

```python
operating = np.array(new_df['Funds_in_million'][new_df['OperatingState']=='Operating'])
```

```python
operating[:20]
```

```python
operating.mean()
```

```python
operating.std()
```

```python
closed = np.array(new_df['Funds_in_million'][new_df['OperatingState']=='Closed'])
```

```python
closed[:5]
```

```python
closed.mean()
```

```python
closed.std()
```

```python
z_statistic, p_value  =  ztest(operating,closed)
print('P Value %1.3f' % p_value)
print('z_statistic %1.3f' % z_statistic)
```

```python
if(p_value <  alpha):
  print("Reject Null Hypothesis")
else:
  print("Fail to Reject NUll Hypothesis")
```

The p value of 0.255 is greater than 5% level of significance.
Hence we fail to reject the null hypothesis.
There is no significant difference between funds raised by operating and closed companies.
