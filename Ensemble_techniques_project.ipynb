```python
# C. Create a function that will accept a DataFrame as input and return pie-charts for all the appropriate Categorical features. Clearly show percentage
# distribution in the pie-chart. [4 Marks]
def plot_pie_chart (dataframe):
    
    for feature in dataframe.columns:
        if dataframe[feature].dtype=='object':
            labels = dataframe[feature].unique()
            sizes = dataframe[feature].value_counts()
                 
            fig1, ax1 = plt.subplots(figsize=(2,2))
            plt.title(feature)
            ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
            ax1.axis('equal')
            plt.show()

```

#### 3. Madel building and Improvement [10 Marks]

```python
# A. Train a model using XGBoost. Also print best performing parameters along with train and test performance. [5 Marks]
import xgboost as xgb
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
from xgboost import XGBClassifier
D_train= xgb.DMatrix(X_train,label=y_train)
D_test = xgb.DMatrix(X_test,label=y_test)
```

```python
param1={
    'eta':0.3,
    'max_depth':5,
    'objective':'multi:softmax',
    'num_class':2
}
steps=20
```

```python
model= xgb.train(param1,D_train,steps)
```

```python
from sklearn.metrics import precision_score,recall_score,accuracy_score
pred= model.predict(D_test)
```

```python
print("Precision={}" .format(precision_score(y_test, pred)))
print("Recall={}" .format(recall_score(y_test, pred)))
print("Accuracy={}" .format(accuracy_score(y_test, pred)))
```

The base XGBoost model gives an  accuracy of about 79%

```python
result_xgb=pd.DataFrame(random_search.cv_results_ )
result_xgb.head()
```

```python
def import_libraries():    # Import the necessary libraries
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    from sklearn.impute import SimpleImputer
    from sklearn.preprocessing import StandardScaler
    from sklearn.model_selection import train_test_split
```

```python
def load_dataset(input_file): # Load the dataset
    global df
    df=pd.read_csv(input_file, index_col='customerID')
    print("\nload_dataset function run successfully!!! \n")
```

```python
def eda_on_data():  #Exploratory data analysis.
    #global df
    print("Snapshot of the  dataset\n",df.head())
    print("Shape of the data:",df.shape)
    print("Columns in data: \n", df.columns)
    print("Description of data: \n", df.describe())
    print("Information on datatypes and null values: \n", df.info())
    print("\n eda on dataset  function run successfully!!! \n")
    
```

```python
def preprocessing():
    global df      # Making the dataframe a global variable as it has to be accessed throughout the program.
    df['TotalCharges'].replace(' ', np.nan, inplace=True) #convert space to null values so we can drop those rows.
    df.dropna(inplace=True)
    print("Null values after imputing",df['TotalCharges'].str.isspace().sum())
    
    df['TotalCharges']= df['TotalCharges'].astype(str).astype(float) # Convert TotalCharges to float.
    df.replace(('Yes','No'),(1,0), inplace=True)
    df.replace(('No phone service','No internet service'),(0,0), inplace=True)
    df=pd.get_dummies(df, columns=['Contract','PaymentMethod'],drop_first=True)
    # Normalizing and scaling the dataframe
    std_scale= StandardScaler()
    std_scale
    df['MonthlyCharges']= std_scale.fit_transform(df[['MonthlyCharges']])
    df['TotalCharges']= std_scale.fit_transform(df[['TotalCharges']])
    #df.drop('customerID', axis=1, inplace=True)
    print("Final preprocessed dataset", df.head)
    print("Data description", df.describe())
    print("\n \n pre processing function run successfully!!!\n ")
    
    
```

```python
def splitting_data():
    global df
    global X, y, X_train,X_test,y_train,y_test
    X= df.drop(labels="Churn", axis=1)
    y=df['Churn']
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=1)
    print("\n \n Splitting data function run successfully!!! \n")
   
```

```python
def train_models():
    from sklearn import svm
    from sklearn.ensemble import RandomForestClassifier
    from sklearn.linear_model import LogisticRegression
    from sklearn.model_selection import GridSearchCV
    from sklearn.neighbors import KNeighborsClassifier
    from sklearn.naive_bayes import GaussianNB
# We train multiple models and then find the best performing model.
    model_params = {
        'svm': {
            'model': svm.SVC(gamma='auto'),
            'params' : {
                'C': [0.01,0.1,1,5,10,15,20],
                'kernel': ['rbf','linear']
            }  
        },
        'random_forest': {
            'model': RandomForestClassifier(),
            'params' : {
                'n_estimators': [1,3,5,7,10]
            }
        },
        'KNN': {
            'model': KNeighborsClassifier(),
            'params' : {
                'n_neighbors': [1,5,10,15,20]
            }
        },
        'Naive Bayes': {
            'model': GaussianNB(),
            'params' : {
                
            }
        },        
        'logistic_regression' : {
            'model': LogisticRegression(solver='liblinear',multi_class='auto'),
            'params': {
               
            }
        }
    }
    
    scores = []

    for model_name, mp in model_params.items():
        clf =  GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=True)
        clf.fit(X,y)
        scores.append({
            'model': model_name,
            'best_score': clf.best_score_,
            'best_params': clf.best_params_,
        })
    print("Best model:", clf.best_score_)
    global score_df
    score_df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
    print(score_df)
    print("\n\n Train models function run successfully!!! \n\n")
```

```python
def best_model(): # Choose the best performing model and create the pickle file for it
    from sklearn import svm
    global score_df,X_train,y_train,X_test,y_test
    param=score_df[score_df.best_score == score_df.best_score.max()]['best_params']
    print(score_df[score_df.best_score == score_df.best_score.max()])
    final_model= svm.SVC(C=5, kernel='rbf')
    final_model.fit(X_train,y_train)
    y_pred= final_model.predict(X_test)
    print(final_model.score(X_test,y_test))
    import pickle
    with open('model_pickle','wb') as f:
        pickle.dump(final_model,f)
        print("\n\n Pickle file created sucessfully!!!")
    print("\n\n END OF PROGRAM !!!")
```

```python
#  Main function
def Main_Function(input_file):
    import_libraries()
    #load_dataset(input_file)
    global df
    global X, y
    df=pd.read_csv(input_file, index_col='customerID')
    eda_on_data()
    preprocessing()
    splitting_data()
    train_models()
    best_model()
    
    import pickle
    with open('model_pickle','rb') as f:
        mp= pickle.load(f) 
    return mp
```
